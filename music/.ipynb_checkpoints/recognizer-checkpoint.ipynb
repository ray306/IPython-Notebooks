```python
%pylab inline
from matplotlib.mlab import find
import matplotlib.pyplot as plt
import math
from array import array
from sys import byteorder
import wave
import numpy as np


noteStrings = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]

def smooth(x,window_len=11,window='hanning'):
    x = np.array(x)
    if x.ndim != 1:
        raise(ValueError, "smooth only accepts 1 dimension arrays.")
    if x.size < window_len:
        raise(ValueError, "Input vector needs to be bigger than window size.")
    if window_len<3:
        return x  
    s=np.r_[x[window_len-1:0:-1],x,x[-1:-window_len:-1]]
    #print(len(s))
    if window == 'flat': #moving average
        w=np.ones(window_len,'d')
    else:
        w=eval('np.'+window+'(window_len)')

    y=np.convolve(w/w.sum(),s,mode='valid')
    return y

def noteFromPitch( frequency ):
    noteNum = 12 * (math.log( frequency / 440 )/math.log(2) )
    try:
        return round( noteNum ) + 69
    except:
        return 0

def noteName(noteID):
    distance2C = noteID -60
    return noteStrings[distance2C%12],distance2C//12+4

def Pitch(signal):
    signal = np.fromstring(signal, 'Int16');
    crossing = [math.copysign(1.0, s) for s in signal]
    index = find(np.diff(crossing));
    return round(len(index) *RATE /(2*np.prod(len(signal))))

chunk = 256
RATE = 44100
freqs = []
pitches = []
res=[]

f = wave.open('new.wav', 'r') 
raw_data = f.readframes(f.getnframes())

for i in range(0, f.getnframes() // 10 ):
    i*=10
    data = raw_data[i:i+chunk]

    if len(data)>0:
        snd_data = array('h', data)
    if byteorder == 'big':
        snd_data.byteswap()
    if max(snd_data) > 100 and Frequency>0:
                   
        freqs.append(Frequency)
        pitches.append(noteFromPitch(Frequency))
#         print(noteName(noteFromPitch(Frequency)))
#     else:
#         freqs.append(None)
#         print('...')

plt.plot(smooth(freqs,window_len=6,window='hanning'))
```

```python
# 升降调
change = 1.25
wav = wave.open('001语音原声.wav', 'r') 
datatype = numpy.dtype('<i' + str(wav.getsampwidth()))  # Little endian
raw_data = numpy.fromstring(wav.readframes(wav.getnframes()), datatype)
channels_data = [raw_data[i::wav.getnchannels()] for i in range(wav.getnchannels())]
wav.close()

def tune(data):
    fft_data = numpy.fft.rfft(data)
    new_fft = [fft_data[round(i/change)] for i in range(len(fft_data))]
    return numpy.fft.irfft(new_fft)

def tuneChannel(data):
    ret = []
    chunksize = wav.getframerate() // 8
    for i in range(0, len(data), chunksize):
        chunk = data[i:i+chunksize]
        ret += list(tune(chunk))
    return ret

final_data = []
for j in range(len(channels_data[0])):
    for i in range(len(channels_data)):
        final_data += [channels_data[i][j]]
final_data=tuneChannel(final_data)
final_data = numpy.array(final_data, dtype=datatype).tostring()

new_wav = wave.open('out.wav', 'w')
new_wav.setparams(wav.getparams())
new_wav.writeframes(final_data)
new_wav.close()
```

```python
from numpy.fft import fft, ifft
import math
import wave
import numpy as np
from scipy.signal import hamming  

index1=15000;
frameSize=256;
spf = wave.open('2015-04-11_21_51_26.wav','r');
fs = spf.getframerate();
signal = spf.readframes(-1);
signal = np.fromstring(signal, 'Int16');
index2=index1+frameSize-1;
frames=signal[index1:int(index2)+1]

zeroPaddedFrameSize=16*frameSize;

frames2=frames*hamming(len(frames));   
frameSize=len(frames);

if (zeroPaddedFrameSize>frameSize):
    zrs= np.zeros(zeroPaddedFrameSize-frameSize);
    frames2=np.concatenate((frames2, zrs), axis=0)

fftResult=np.log(abs(fft(frames2)));
ceps=ifft(fftResult);

posmax = ceps.argmax();

result = fs/zeroPaddedFrameSize*(posmax-1)

print(result)
```
