```python
# 升降调
change = 1.25
wav = wave.open('001语音原声.wav', 'r') 
datatype = numpy.dtype('<i' + str(wav.getsampwidth()))  # Little endian
raw_data = numpy.fromstring(wav.readframes(wav.getnframes()), datatype)
channels_data = [raw_data[i::wav.getnchannels()] for i in range(wav.getnchannels())]
wav.close()

def tune(data):
    fft_data = numpy.fft.rfft(data)
    new_fft = [fft_data[round(i/change)] for i in range(len(fft_data))]
    return numpy.fft.irfft(new_fft)

def tuneChannel(data):
    ret = []
    chunksize = wav.getframerate() // 8
    for i in range(0, len(data), chunksize):
        chunk = data[i:i+chunksize]
        ret += list(tune(chunk))
    return ret

final_data = []
for j in range(len(channels_data[0])):
    for i in range(len(channels_data)):
        final_data += [channels_data[i][j]]
final_data=tuneChannel(final_data)
final_data = numpy.array(final_data, dtype=datatype).tostring()

new_wav = wave.open('out.wav', 'w')
new_wav.setparams(wav.getparams())
new_wav.writeframes(final_data)
new_wav.close()
```

```python
from numpy.fft import fft, ifft
import math
import wave
import numpy as np
from scipy.signal import hamming  

index1=15000;
frameSize=256;
spf = wave.open('2015-04-11_21_51_26.wav','r');
fs = spf.getframerate();
signal = spf.readframes(-1);
signal = np.fromstring(signal, 'Int16');
index2=index1+frameSize-1;
frames=signal[index1:int(index2)+1]

zeroPaddedFrameSize=16*frameSize;

frames2=frames*hamming(len(frames));   
frameSize=len(frames);

if (zeroPaddedFrameSize>frameSize):
    zrs= np.zeros(zeroPaddedFrameSize-frameSize);
    frames2=np.concatenate((frames2, zrs), axis=0)

fftResult=np.log(abs(fft(frames2)));
ceps=ifft(fftResult);

posmax = ceps.argmax();

result = fs/zeroPaddedFrameSize*(posmax-1)

print(result)
```
