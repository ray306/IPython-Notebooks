```python
%pylab inline

import random
from PIL import Image
import os
import re
import numba
from time import time

#Create Weight matrix for a single image
#建立向量中每两个点之间的连接
@numba.jit()
def create_W(x):
    w = np.zeros([len(x),len(x)])
    for i in range(len(x)):
        for j in range(i,len(x)):
            if i == j:
                w[i,j] = 0
            else:
                w[i,j] = x[i]*x[j]
                w[j,i] = w[i,j]
    return w

#Read Image file and convert it to Numpy array
def img2array(file,size, threshold= 145):
    pilIN = Image.open(file).convert(mode="L")
    pilIN = pilIN.resize(size)
    imgArray = np.asarray(pilIN,dtype=np.uint8)
    x = np.zeros(imgArray.shape,dtype=np.float)
    x[imgArray > threshold] = 1
    x[x==0] = -1
    return x

#Convert Numpy array to Image file like Jpeg
def array2img(data):
    #data is 1 or -1 matrix
    y = np.zeros(data.shape,dtype=np.uint8)
    y[data==1] = 255
    y[data==-1] = 0
    img = Image.fromarray(y,mode="L")
    return img

#Update
def update(w,y_vec,theta=0.5,time=100):
    for s in range(time):
        m = len(y_vec)
        i = random.randint(0,m-1)
        u = np.dot(w[i][:],y_vec) - theta

        if u > 0:
            y_vec[i] = 1
        elif u < 0:
            y_vec[i] = -1

    return y_vec


#The following is training pipeline
#Initial setting

def hopfield(train_files, test_files,theta=0.5, times=1000, size=(100,100),threshold=60, current_path=None):
    #read image and convert it to Numpy array
    print("Importing images and creating weight matrix....")
    w = np.zeros([size[0]*size[1],size[0]*size[1]])
    #num_files is the number of files
    num_files = 0
    for path in train_files:
        print(path)
        x = img2array(file=path,size=size,threshold=threshold)
        x_vec = x.flatten()
        print(len(x_vec))
        
        w += create_W(x_vec)

    print("Weight matrix is done!!")


    #Import test data
    fig,axes = plt.subplots(len(test_files),2)
    for i,path in enumerate(test_files):
        y = img2array(file=path,size=size,threshold=threshold)
 
        oshape = y.shape
        y_img = array2img(y)
        print("Imported test data")

        y_vec = y.flatten()

        print("Updating...")
        y_vec_after = update(w=w,y_vec=y_vec,theta=theta,time=times)

        y_vec_after = y_vec_after.reshape(oshape)

        after_img=array2img(y_vec_after)
        axes[i][0].imshow(np.asarray(y_img))
        axes[i][1].imshow(np.asarray(after_img))
```
