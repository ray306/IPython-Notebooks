```python
# sub12 marker只有1 4 8, 21 31不是正确与否
import re
import pandas as pd
path = 'F:/eeg/rawdata/sub12/'
resps = pd.read_csv(path+'12_result.csv',encoding='gbk')
def mark_set(x):
    if x[1]==1:
        mark = 'T'
    else:
        mark = 'F'
    if x[2]==1:
        mark += 't'
    else:
        mark += 'f'
    if x[4]==0:
        mark += 'G'
    else:
        mark += 'L'
    
    if (x[7]=='j' and x[5]==1) | (x[7]=='f' and x[5]==0):
        expect = 1
    else:
        expect = 0
        
    return mark,expect

resps = list(resps.apply(lambda x: mark_set(x),axis=1))

with open(path+'sub12.vmrk') as f:
    raw = f.read()
    new = []
    Start = False
    markID = 1
    count = 0
    for l in raw.split('\n'):
        if not re.search(r'Mk.*?=', l):
            new.append(l)
        elif '=New Segment' in l:
            new.append('MK'+str(markID)+'='+l.split('=')[1])
            markID+=1
        elif 'Stimulus' in l:
            time = l.split(',')[2]
            if 'S  ' in l and resps[count][1] == 1:
                typeS = resps[count][0]
                new.append('MK%d=Stimulus,S  %s,%s,1,0' %(markID, typeS, time)) 
                markID+=1
            if 'S 31' in l or 'S 21' in l:
                if resps[count][1] == 1:
                    typeS = resps[count][1]
                    new.append('MK%d=Stimulus,S %s,%s,1,0' %(markID, typeS, time)) 
                    markID+=1
                count+=1
                
with open(path+'sub12.vmrk','w+') as f:
    f.write('\n'.join(new))
```

```python
import pandas as pd
path = 'F:\\eeg\\rawdata\\sub11\\'
Type = {
    '000':'FfL',
    '001':'FfG',
    '010':'FtL',
    '011':'FtG',
    '100':'TfL',
    '101':'TfG',
    '110':'TtL',
    '111':'TtG'
        }
for num in range(1,10):
    file1 = path+'sub11b'+str(num)
    file2 = path+'11_result.csv'
    behavior = pd.read_csv(file2,encoding='gbk')
    behavior['Acc'] = behavior[['Expected','respKey']].apply(lambda x:True if (x[0]==1 and x[1]=='j') or (x[0]==0 and x[1]=='f') else False, axis=1)
    behavior['Type'] = behavior[['WholeAcc','LeftAcc','Pos']].apply(lambda x:''.join([str(x[0]),str(x[1]),str(int(x[2]==0))]),axis=1)

    behavior = behavior[60*num-60:60*num]
    with open(file1+'.vmrk') as f:
        raw = f.read()
        new = []
        marker = []

        for l in raw.split('\n'):
            if l!='':
                if not 'Stimulus,S' in l:
                    new.append(l)
                else:
                    marker.append(l)
        count = int(marker[0][2])        
        trials = [marker[i:i+2] for i in range(0,len(marker),2)]


        for (idx,trial),(m1,m2) in zip(behavior.iterrows(),trials):
            if trial.Acc == True:
                m1 = m1.replace('S  1','S  '+Type[trial.Type])
                new.append('Mk'+str(count)+'='+m1.split('=')[1])
                count+=1
                m2 = m2.replace('S 20','S 11').replace('S 21','S 11')
                new.append('Mk'+str(count)+'='+m2.split('=')[1])
                count+=1

    with open(file1+'.vmrk','w+') as f:
        f.write('\n'.join(new)) 
```
