```python
import pandas as pd
import random
%pylab inline

rawdata = pd.read_csv('d:\\bi_frame_filtered3.csv',encoding='utf-8')
del rawdata['Unnamed: 0']
rawdata.head()
```

```python
n=len(rawdata)//2
index = list(range(n*2))
random.shuffle(index)
index=index[:n//4]
n=n//4
group_holdW1 = rawdata.ix[index[:n]]
group_holdW2 = rawdata.ix[index[n:]]

block=[[] for i in range(4)]

seq = createSeq(n)
for idx,i in enumerate(group_holdW1.iterrows()):#检测词在左
    now = i[1]
    for blockID,stimType in zip(range(4),seq[idx]):
        if stimType==1:
            stim = (now.w1,now.w2),now.w1,'1','1','l'
        elif stimType==2:
            stim = (now.w1,now.w2),now.w1_pseudo1,'1','0','l'
        elif stimType==3:
            stim = (now.w1,now.w2_match),now.w1,'0','1','l'
        elif stimType==4:
            stim = (now.w1,now.w2_match),now.w1_pseudo2,'0','0','l'
        block[blockID].append([str(i[0]),stim,str(now.cooccur),str(now.freq1),str(now.freq2)])
    
seq = createSeq(n)
for idx,i in enumerate(group_holdW2.iterrows()):#检测词在右
    now = i[1]
    for blockID,stimType in zip(range(4),seq[idx]):
        if stimType==1:
            stim = (now.w1,now.w2),now.w2,'1','1','r'
        elif stimType==2:
            stim = (now.w1,now.w2),now.w2_pseudo1,'1','0','r'
        elif stimType==3:
            stim = (now.w1_match,now.w2),now.w2,'0','1','r'
        elif stimType==4:
            stim = (now.w1_match,now.w2),now.w2_pseudo2,'0','0','r'
        block[blockID].append([str(i[0]),stim,str(now.cooccur),str(now.freq1),str(now.freq2)])

for i in range(4):
    with open('D:\\Coding\\py\\py3\\experiment\\stims_block'+str(i+1)+'.csv','w',encoding='utf-8') as f:
        f.write(',prime,target,bigram,targetIsTrue,targetPos,cooccur,freq1,freq2\n')
        random.shuffle(block[i])
        for stim in block[i]:
            f.write(','.join([stim[0],stim[1][0][0]+stim[1][0][1],stim[1][1],stim[1][2],stim[1][3],stim[1][4],stim[2],stim[3],stim[4]])+'\n')
            

with open('D:\\Coding\\py\\py3\\experiment\\stims_practice.csv','w',encoding='utf-8') as f:
    f.write(',prime,target,bigram,targetIsTrue,targetPos,cooccur,freq1,freq2\n')
    
    practice = []
    seq = createSeq(n)
    for idx,i in enumerate(group_holdW2.iterrows()):#检测词在左
        now = i[1]
        for stimType in seq[idx]:
            if stimType==1:
                stim = (now.w1,now.w2),now.w1,'1','1','l'
            elif stimType==2:
                stim = (now.w1,now.w2),now.w1_pseudo1,'1','0','l'
            elif stimType==3:
                stim = (now.w1,now.w2_match),now.w1,'0','1','l'
            elif stimType==4:
                stim = (now.w1,now.w2_match),now.w1_pseudo2,'0','0','l'
            practice.append(','.join([str(i[0]),stim[0][0]+stim[0][1],stim[1],stim[2],stim[3],stim[4],str(now.cooccur),str(now.freq1),str(now.freq2)])+'\n')

    seq = createSeq(n)
    for idx,i in enumerate(group_holdW1.iterrows()):#检测词在右
        now = i[1]
        for stimType in seq[idx]:
            if stimType==1:
                stim = (now.w1,now.w2),now.w2,'1','1','r'
            elif stimType==2:
                stim = (now.w1,now.w2),now.w2_pseudo1,'1','0','r'
            elif stimType==3:
                stim = (now.w1_match,now.w2),now.w2,'0','1','r'
            elif stimType==4:
                stim = (now.w1_match,now.w2),now.w2_pseudo2,'0','0','r'
            practice.append(','.join([str(i[0]),stim[0][0]+stim[0][1],stim[1],stim[2],stim[3],stim[4],str(now.cooccur),str(now.freq1),str(now.freq2)])+'\n')

    random.shuffle(practice)
    for i in practice[:20]:
        f.write(i)
```
