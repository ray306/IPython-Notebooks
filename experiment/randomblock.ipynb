```python
import random
import numpy as np
pos = [0,0,0,0,1,1,2,2]
random_pos = np.zeros((4,len(pos)),dtype=int)
answers = [0,0,0,0,1,1,1,1]
random_ans = np.zeros((4,len(answers)),dtype=int)

while(1):
    for i in range(4):
        np.random.shuffle(pos)
        random_pos[i] = pos

    for i in range(4):
        np.random.shuffle(answers)
        random_ans[i] = answers

    Pos = random_pos.flatten()
    Ans = random_ans.flatten()
    
    com = pd.DataFrame([str(i[0])+str(i[1]) for i in zip(Pos,Ans)])
    if len(com[com[0]=='11'])==len(com[com[0]=='10']) and len(com[com[0]=='01'])==len(com[com[0]=='00']) and len(com[com[0]=='21'])==len(com[com[0]=='20']):
        stimli['Pos'] = random_pos.flatten()
        stimli['Ans'] = random_ans.flatten()
        break
```
